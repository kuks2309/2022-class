```python
import matplotlib.pyplot as plt
import numpy as np
from random import random
```

```python
data = [ [4,6], [8,5], [12,15], [16,12], [20,20] ]

Data = np.array(data)
X = np.array(Data[:,0])
Y = np.array(Data[:,1])
X1 = np.zeros(5)
Y1 = np.zeros(5)
print(X)
```

```python
x_sum = X.sum()
y_sum = Y.sum()
x_avg = X.mean()
y_avg = Y.mean()

print(X[0])
```

```python
sum_xy  = 0.0
sum_xx2 = 0.0

for i in range(0, len(X)):
    sum_xy  += (X[i] - x_avg) * (Y[i] - y_avg)
    sum_xx2 += (X[i] - x_avg) * (X[i] - x_avg)

print(x_sum, y_sum,x_avg,y_avg)
print(sum_xy, sum_xx2)

```

```python
a = sum_xy/sum_xx2
b = y_avg - a*x_avg
```

```python
for i in range(0,len(X)):
    Y1[i] = X[i]*a + b
```

```python
print(a,b)
```

```python
plt.figure("Linear Regression")
plt.scatter(X,Y)
#plt.plot(X,Y, 'g--', X1,Y1, c='r')
plt.plot(X,Y1, c='r')
plt.show()
```
