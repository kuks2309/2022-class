```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import io
```

```python
input_mat = io.loadmat('/content/SonarAl.mat')
```

```python
print(input_mat)
```

```python
print(input_mat['sonarAlt'])
print(input_mat['sonarAlt'][0])
data_sonar = input_mat['sonarAlt'][0]
print(data_sonar)
print(len(data_sonar))
```

```python
def get_sonar(i):
    """Measure sonar."""
    z = input_mat['sonarAlt'][0][i]  # input_mat['sonaralt']: (1, 1501)
    return z
```

```python
def mov_avg_filter(x_n, x_meas):
    """Calculate average sonar using a moving average filter (batch expression)."""
    n = len(x_n)
    for i in range(n-1):
        x_n[i] = x_n[i+1]
    x_n[n-1] = x_meas
    x_avg = np.mean(x_n)
    return x_avg, x_n
```

```python
# Input parameters.
n = 10
n1 = 20
n_samples = 500
time_end = 20
```

```python
dt = time_end / n_samples
time = np.arange(0, time_end, dt)
x_meas_save = np.zeros(n_samples)
x_avg_save = np.zeros(n_samples)
x_meas_save_20 = np.zeros(n_samples)
x_avg_save_20 = np.zeros(n_samples)
```

```python
print(time)
print(x_meas_save)
print(x_avg_save)
```

```python
for i in range(n_samples):
    x_meas = get_sonar(i)
    if i == 0:
        x_avg, x_n = x_meas, x_meas * np.ones(n)
        #print(x_meas * np.ones(n))
    else:
        x_avg, x_n = mov_avg_filter(x_n, x_meas)
        #print(x_n)   
    x_meas_save[i] = x_meas
    x_avg_save[i] = x_avg
```

```python
for i in range(n_samples):
    x_meas = get_sonar(i)
    if i == 0:
        x_avg_20, x_n1 = x_meas, x_meas * np.ones(n1)
        #print(x_meas * np.ones(n1))
    else:
        x_avg_20, x_n1 = mov_avg_filter(x_n1, x_meas)
        #print(x_n)   
    x_meas_save_20[i] = x_meas
    x_avg_save_20[i] = x_avg_20
```

```python
plt.plot(time, x_meas_save, 'r*', label='Measured')
plt.plot(time, x_avg_save,    'b-', label='Moving average_10')
plt.plot(time, x_avg_save_20, 'g-', label='Moving average_20')
plt.legend(loc='upper left')
plt.title('Measured Altitudes v.s. Moving Average Filter Values')
plt.xlabel('Time [sec]')
plt.ylabel('Altitude [m]')
plt.savefig('png/moving_average_filter.png')
```
